```python
# Read Fashion MNIST dataset
import util_mnist_reader
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame as df
from keras.utils.np_utils import to_categorical
from sklearn import preprocessing, metrics
from matplotlib import pyplot
# from sklearn import s
X_train, y_train = util_mnist_reader.load_mnist('data/fashion', kind='train')
X_test, y_test = util_mnist_reader.load_mnist('data/fashion', kind='t10k')

X_train = X_train/255
X_test = X_test/255

y_train = to_categorical(y_train, num_classes=10)# one-hot vectors
y_test = to_categorical(y_test, num_classes=10) 
```

```python
def Sigmoid(Z):
    return (1/ (1+np.exp(-Z)))

def Softmax(Z):
    den = (np.sum(np.exp(Z), axis = 1))
    den = np.matrix(den.transpose())
    num = (np.exp(Z)).transpose()
    return (num/den).transpose()
```

```python
#hyperparameters

alpha = 0.30 # Learning Rate
n = 100 # No. of neurons in 2nd layer
epochs = 1000


#Weights between input and hidden layer
W1 = np.random.normal(scale=0.1, size=(X_test.shape[1],n))
b1 = np.random.normal(scale=0.1, size=(n,1))
#Weights between hidden and output layer
W2 = np.random.normal(scale=0.1, size=(n,10))
b2 = np.random.normal(scale=0.1, size=(10,1))

loss_plot = list()
val_loss_plot = list()


A0 = X_train # Input layer
Z1 = np.dot(A0,W1).transpose() + b1
A1 = Sigmoid(Z1.transpose()) # Hidden layer
Z2 = np.dot(A1,W2).transpose() + b2
A2 = Softmax(Z2.transpose()) # output layer

for epoch in range(epochs):
    dW2 = np.dot(A1.transpose(),(A2-y_train))/60000
    dL_dZ2 = (A2-y_train)
    dZ2_dA1 = W2


    dL_dA1 = np.dot(dL_dZ2, dZ2_dA1.transpose()) #60k,100
    dA1_dZ1 = np.multiply(A1,1-A1) #60k,100
    dZ1_dW1 = X_train

    dL_dZ1 = np.multiply(dL_dA1, dA1_dZ1)
    dL_dW1 = np.dot(dZ1_dW1.transpose(),dL_dZ1)
    dW1 = dL_dW1/60000
    
    db1 = np.sum(dL_dZ1,axis=0)/60000
    db1 = db1.transpose()
    db2 = np.sum(dL_dZ2,axis=0)/60000
    db2 = db2.transpose()
       
    
    W2 = W2 - (alpha * dW2)
    W1 = W1 - (alpha * dW1)
    
    b1 = b1 - (alpha * db1)
    b2 = b2 - (alpha * db2)
    
    #print (b1,b2)


    Z1 = np.dot(A0,W1).transpose() + b1
    A1 = Sigmoid(Z1.transpose()) # Hidden layer
    Z2 = np.dot(A1,W2).transpose() + b2
    A2 = Softmax(Z2.transpose()) # output layer
    
    loss = - np.sum(np.multiply(y_train,np.log(A2)))/60000
    
    
    
    pA0 = X_test # Input layer
    pZ1 = np.dot(pA0,W1).transpose() + b1
    pA1 = Sigmoid(pZ1.transpose()) # Hidden layer
    pZ2 = np.dot(pA1,W2).transpose() + b2
    pA2 = Softmax(pZ2.transpose()) # output layer
    
    val_loss = - np.sum(np.multiply(y_test, np.log(pA2)))/10000
    print("Epoch",epoch,"\t","Loss:",loss)
    
    loss_plot.append(loss)
    val_loss_plot.append(val_loss)

pA0 = X_test # Input layer
pZ1 = np.dot(pA0,W1).transpose() + b1
pA1 = Sigmoid(pZ1.transpose()) # Hidden layer
pZ2 = np.dot(pA1,W2).transpose() + b2
pA2 = Softmax(pZ2.transpose()) # output layer


actual = y_test.argmax(axis=1)
prediction = pA2.argmax(axis=1)
pyplot.plot(loss_plot,label="Training Loss")
pyplot.plot(val_loss_plot,label="Validation Loss")
pyplot.legend()
pyplot.xlabel("epochs")
pyplot.ylabel("loss")
accuracy = metrics.accuracy_score(actual,prediction)

print (accuracy)
```

```python
pA0 = X_test # Input layer
pZ1 = np.dot(pA0,W1).transpose() + b1
pA1 = Sigmoid(pZ1.transpose()) # Hidden layer
pZ2 = np.dot(pA1,W2).transpose() + b2
pA2 = Softmax(pZ2.transpose()) # output layer


actual = y_test.argmax(axis=1)
prediction = pA2.argmax(axis=1)
metrics.confusion_matrix(actual,prediction)
```

# Task 2

```python
# Read Fashion MNIST dataset

import util_mnist_reader
X_train, y_train = util_mnist_reader.load_mnist('data/fashion', kind='train')
X_test, y_test = util_mnist_reader.load_mnist('data/fashion', kind='t10k')

import keras
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame as df
from keras import Sequential
from keras import layers
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(128, activation="relu", input_dim = 784))
model.add(Dense(128, activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(10,activation="softmax"))

adm = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss=keras.losses.sparse_categorical_crossentropy,
              optimizer=adm, metrics =["accuracy"])

history = model.fit(X_train, y_train, epochs=15,validation_data=(X_test,y_test))

plt.plot(history.history["loss"], label = "Training loss")
plt.plot(history.history["val_loss"],label="validation Loss")
plt.title("Loss")
plt.xlabel("Epocs")
plt.ylabel("Loss")
plt.legend()

score = model.evaluate(X_test, y_test)
accuracy = score[1]
print ("Accuracy",accuracy)
```

```python
prediction = model.predict(X_test)
prediction = prediction.argmax(axis=1)

metrics.confusion_matrix(y_test,prediction)
```

```python
import util_mnist_reader
X_train, y_train = util_mnist_reader.load_mnist2('data/fashion', kind='train')
X_test, y_test = util_mnist_reader.load_mnist2('data/fashion', kind='t10k')

import keras
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame as df

from keras import Sequential
from keras import layers
from keras.layers import Dense
```

```python
#Normalization
X_train = (X_train/ 255)
# X_train = X_train.reshape(60000,28,28)
X_test = X_test/ 255
```

```python
model = Sequential()

model.add(layers.Conv2D(filters=64, kernel_size=2, activation='relu', input_shape=(28,28,1))) 
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Conv2D(filters=32, kernel_size=2, activation='relu'))
model.add(layers.MaxPooling2D(pool_size=2))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

adm = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='sparse_categorical_crossentropy', optimizer=adm,  metrics=['accuracy'])

history = model.fit(X_train, y_train, batch_size=64, epochs=10,validation_data=(X_test,y_test))

plt.plot(history.history["loss"], label = "Training loss")
plt.plot(history.history["val_loss"],label="validation Loss")
plt.title("Loss")
plt.xlabel("Epocs")
plt.ylabel("Loss")
plt.legend()

score = model.evaluate(X_test, y_test)
accuracy = score[1]
print (accuracy)
```

```python
prediction = model.predict(X_test)
prediction = prediction.argmax(axis=1)

metrics.confusion_matrix(y_test,prediction)
```
