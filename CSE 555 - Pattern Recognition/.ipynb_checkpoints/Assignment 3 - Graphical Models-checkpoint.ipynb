```python
from pgmpy.models import BayesianModel
import networkx as nx
import pylab as plt
```

```python
graph = BayesianModel()
nodes = ["asia", "tub", "smoke", "lung", "either", "bronc", "dysp", "xray" ]
edges = [("asia", "tub"), ("tub", "either"), ("either", "xray"), ("smoke", "lung"), ("smoke", "bronc"), ("bronc", "dysp"), ("either", "dysp"), ("lung", "either")]

graph.add_nodes_from(nodes)
graph.add_edges_from(edges)
```

#### Bayesian Network

```python
nx.draw(graph, with_labels=True)
print("Bayesian Network")
plt.show()
```

#### Moral Graph

```python
# Moral Graph
moral_graph = graph.to_markov_model()
nx.draw(moral_graph, with_labels=True)
print("Moral Graph")
plt.show()
```

#### Triangulated Graph

```python
traing_graph = moral_graph.triangulate()
nx.draw(traing_graph, with_labels=True)
print("Traingulated Graph")
plt.show()
```

#### Junction Tree

```python
junc_tree = graph.to_junction_tree()
print (junc_tree)
nx.draw(junc_tree, with_labels=True)
print("Junction Tree")
img = plt.show()
```

```python
from pgmpy.inference import BeliefPropagation
from pgmpy.factors.discrete import TabularCPD

asia_CPD = TabularCPD(variable="asia", variable_card=2, values=[[0.01, 0.99]])
tub_CPD = TabularCPD(variable="tub", variable_card=2, values=[[0.05, 0.01],[0.95, 0.99]], evidence_card=[2], evidence=["asia"])
smoke_CPD = TabularCPD(variable="smoke", variable_card=2, values=[[0.5, 0.5]])
lung_CPD = TabularCPD(variable="lung", variable_card=2, values=[[0.1, 0.01],[0.9, 0.99]], evidence_card=[2], evidence=["smoke"])
either_CPD = TabularCPD(variable="either", variable_card=2, values=[[1, 1, 1, 0],[0, 0, 0, 1]], evidence_card=[2,2], evidence=["tub", "lung"])
bronc_CPD = TabularCPD(variable="bronc", variable_card=2, values=[[0.6, 0.3],[0.4, 0.7]], evidence_card=[2], evidence=["smoke"])
dysp_CPD = TabularCPD(variable="dysp", variable_card=2, values=[[0.9, 0.7, 0.8, 0.1],[0.1, 0.3, 0.2, 0.9]], evidence_card=[2,2], evidence=["either", "bronc"])
xray_CPD = TabularCPD(variable="xray", variable_card=2, values=[[0.98, 0.05], [0.02, 0.95]], evidence_card=[2], evidence=["either"] )

graph.add_cpds(asia_CPD, tub_CPD, smoke_CPD, lung_CPD, either_CPD, bronc_CPD, dysp_CPD, xray_CPD )

print (graph.nodes())
print ("ASIA")
print(graph.get_cpds("asia"))
print ("TUB")
print(graph.get_cpds("tub"))
print ("SMOKE")
print(graph.get_cpds("smoke"))
print ("LUNG")
print(graph.get_cpds("lung"))
print ("EITHER")
print(graph.get_cpds("either"))
print ("BRONC")
print(graph.get_cpds("bronc"))
print ("DYSP")
print(graph.get_cpds("dysp"))
print ("XRAY")
print(graph.get_cpds("xray"))
```

```python
belief_propagation = BeliefPropagation(graph)
result = belief_propagation.query(variables=["tub", "lung", "bronc"], evidence={"asia":0, "xray":0})
print(result)
```

```python
from pgmpy.inference import VariableElimination
variable_elimination = VariableElimination(graph)
result = variable_elimination.query(variables=["tub", "lung", "bronc"], evidence={"asia":0, "xray":0})
print(result)
```

Therefore from above table, the Joint Probability of "tub=yes, lung=yes, bronc=yes", given evidence that "asia=yes, xray=yes" is 0.0106

 #### Explain why the "running intersection property" is satisfied in your junction tree.

Running intersection property states that if a variable found in 2 clusters, then that variable should be in the path that connects two clusters

In my junction tree below,
- "either" is present in all clusters 1,4 so it also appears in all intermediate clusters (c2, c3)
- "dysp" is present in cluster 2,4 so it also appears in all intermediate clusters(c3)
- "xray" is seen only in cluster 1  but doesnt present in cluster 2 so it is not present in any other cluster
