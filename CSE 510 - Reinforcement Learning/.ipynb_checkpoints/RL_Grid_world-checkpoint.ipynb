```python
import gym
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Deterministic Environment - 5x5 Grid World

- **Actions**
    * 0 : up
    * 1 : down
    * 2 : left
    * 3 : right

- **Rewards**
    * Terminal State : +10
    * Gold Coin : +1.5
    * Diamond : +3
    * Any enemy : -5

```python
class Grid5x5_det():
    
    def __init__(self):
        self.action_space = [0,1,2,3]
        self.observation_space = list()
        for i in range(0,5):
            for j in range(0,5):
                self.observation_space.append((i,j))


        self.coin = [0,4]
        self.diamond = [3,1]
        self.enemies = [[1,2]]        
        self.terminal_state = [4,4]
        
    def _get_distance(self, x, y):
        return abs(x[0] - y[0]) + abs(x[1] - y[1])

    def step(self, action, state):
        dis_1 = self._get_distance(state ,self.terminal_state)
        next_state = state.copy()
        if action == 0:
            if state[0] > 0:
                print ("moving up")
                next_state[0] -= 1        
        elif action == 1:
            if state[0] < 4:
                print ("moving down")
                next_state[0] += 1
        elif action == 2:
            if state[1] > 0:
                print ("moving left")
                next_state[1] -= 1
        elif action == 3:
            if state[1] < 4:
                print ("moving right")
                next_state[1] += 1

        self.reward = 0
        self.status = False
        
        dis_2 = self._get_distance(next_state,self.terminal_state)

        print ("current distance is ",dis_1, "future distance is ",dis_2)
        
        if next_state == self.terminal_state:
            self.reward = 10
            self.status = True

        elif next_state in self.enemies:
            self.reward = -5
        elif next_state == self.diamond:
            self.reward = 3
            self.diamond = [100,100]
        elif next_state == self.coin:
            self.reward = 1.5
            self.coin = [100,100]

        elif dis_2 >= dis_1:
            self.reward = self.reward -1
        
        elif dis_2 < dis_1:
            self.reward += 1
        

        
        return next_state, self.reward, self.status
            
    
    
    def reset(self):
        self.coin = [0,4]
        self.diamond = [3,1]
        self.enemies = [[1,2]]        
        self.terminal_state = [4,4] 
        
    def render(self,next_state):
        self.render_matrix = np.zeros((5,5,3),int) + (50,50,50)
        for [a,b] in self.enemies:
            self.render_matrix[a,b] = (178,34,34)
        self.render_matrix[3,1] = (100,149,237) # diamond
        self.render_matrix[0,4] = (255,215,0) # coin
        self.render_matrix[4,4] = (46,139,87) # terminal
        self.render_matrix[next_state[0],next_state[1]] = (200,200,200)
        plt.imshow(self.render_matrix)
            
```

```python
env = Grid5x5_det()
env.render((0,0))
```

## Q Learning - Deterministic

```python
env = Grid5x5_det()

episodes = 999
epsilon = 1
q_table = np.zeros((5,5,4))
alpha = 0.7
gamma = 0.7

epsilon_history = list()
total_reward_history = list()


for episode in range(0, episodes):
    print (epsilon, episode)
    epsilon = np.exp(-0.008*episode)
    epsilon_history.append(epsilon)
    state = [0,0]
    itera = 0
    total_reward  = 0
    env.reset()
    status = False
    
    while(not status):
        if (itera>20):
            total_reward_history.append(total_reward)
            break
        print ("Episode - ",episode)
        random_num = random.randint(0,100)
        if random_num > epsilon*100:
            action = np.argmax(q_table[(state[0],state[1])])
            next_state, reward, status = env.step(action,state)

        else:
            action = random.randint(0,3)
            next_state, reward, status = env.step(action,state)

        if (episode == 998):
            env.render(next_state)
        
        q_table[state[0]][state[1]][action] = (1-alpha) * q_table[state[0]][state[1]][action] + alpha * (reward + gamma* np.max( q_table[next_state[0]][next_state[1]][:]) )
        total_reward += reward
    
        state = next_state
        itera += 1
        plt.show()
        
        if state==[4,4]:
            total_reward_history.append(total_reward)
            
            print (total_reward)
            print ("-"*250)
            
    
```

```python
#Total Reward History
plt.plot(total_reward_history)
```

```python
plt.plot(total_reward_history,epsilon_history)
```

## Stochastic Environment - 5x5 Grid World

- **Actions**
    * 0 : up
    * 1 : down
    * 2 : left
    * 3 : right

- **Rewards**
    * Terminal State : +7  
    * Gold Coin : +2
    * Diamond : +5
    * Any enemy : -3
<br>
<br>
* Model has 10 % chance for reaching opposite state and 5 % chance of reaching other state
* Example if action = up -> 85 % chance to go up, 10 % chance to go down, 5% chance to go left/right
* Example if action = left -> 85 % chance to go left, 10 % chance to go right, 5% chance to go up/down


```python
class Grid5x5_stoc():
    
    def __init__(self):
        self.action_space = [0,1,2,3]
        self.observation_space = list()
        for i in range(0,5):
            for j in range(0,5):
                self.observation_space.append((i,j))


        self.coin = [0,4]
        self.diamond = [3,1]
        self.enemies = [[1,2]]        
        self.terminal_state = [4,4]
        
    def _get_distance(self, x, y):
        return abs(x[0] - y[0]) + abs(x[1] - y[1])

    def step(self, action, state):
        dis_1 = self._get_distance(state ,self.terminal_state)
        next_state = state.copy()
                # UP
        if action == 0:
            rand = random.randint(0,100)
            # print (rand)
            if rand < 85 :
                if self.state[0] > 0:
                    print ("moving up")
                    self.next_state[0] -= 1
                else:
                    print ("cant move up")
            elif rand <95:
                self.step(1)
            else:
                self.step(random.randint(2,3))
                
        # DOWN
        elif action == 1:
            rand = random.randint(0,100)
            if rand < 85 :
                if self.state[0] < 4:
                    print ("moving down")
                    self.next_state[0] += 1
                else:
                    print ("cant move down")
            
            elif rand <95:
                self.step(0)
            else:
                self.step(random.randint(2,3))
        
        # Left
        elif action == 2:
            rand = random.randint(0,100)
            if rand < 85 :
                if self.state[1] > 0:
                    print ("moving left")
                    self.next_state[1] -= 1
                else:
                    print ("cant move left")
            
            elif rand <95:
                self.step(3)
                
            else:
                self.step(random.randint(0,1))
            
         #RIGHT
        elif action == 3:
            rand = random.randint(0,100)
            if rand < 85 :
                if self.state[1] < 4:
                    print ("moving left")
                    self.next_state[1] += 1
                else:
                    print ("cant move up")
            elif rand <95:
                self.step(2)
            else:
                self.step(random.randint(0,1))

        self.reward = 0
        self.status = False
        
        dis_2 = self._get_distance(next_state,self.terminal_state)

        print ("current distance is ",dis_1, "future distance is ",dis_2)
        
        if next_state == self.terminal_state:
            self.reward = 10
            self.status = True

        elif next_state in self.enemies:
            self.reward = -5
        elif next_state == self.diamond:
            self.reward = 3
            self.diamond = [100,100]
        elif next_state == self.coin:
            self.reward = 1.5
            self.coin = [100,100]

        elif dis_2 >= dis_1:
            self.reward = self.reward -1
        
        elif dis_2 < dis_1:
            self.reward += 1
        

        
        return next_state, self.reward, self.status
            
    
    
    def reset(self):
        self.coin = [0,4]
        self.diamond = [3,1]
        self.enemies = [[1,2]]        
        self.terminal_state = [4,4] 
        
    def render(self,next_state):
        self.render_matrix = np.zeros((5,5,3),int) + (50,50,50)
        for [a,b] in self.enemies:
            self.render_matrix[a,b] = (178,34,34)
        self.render_matrix[3,1] = (100,149,237) # diamond
        self.render_matrix[0,4] = (255,215,0) # coin
        self.render_matrix[4,4] = (46,139,87) # terminal
        self.render_matrix[next_state[0],next_state[1]] = (200,200,200)
        plt.imshow(self.render_matrix)
            
```

## Q Learning - Stochastic

```python
env2 = Grid5x5_stoc()

episodes = 1000
epsilon = 1
q_table = np.zeros((5,5,4))
total_reward_history = list()
epsilon_history = list()
alpha = 0.7
gamma = 0.7


for episode in range(0, episodes):
    epsilon_history.append(epsilon)
    print (epsilon, episode)
    epsilon = np.exp(-0.008*episode)

    state = [0,0]
    itera = 0
    total_reward  = 0
    env2.reset()
    status = False
    
    while(not status):
        if (itera>20):
            total_reward_history.append(total_reward)
            break
        print (episode,epsilon)
        random_num = random.randint(0,100)
        if random_num > epsilon*100:
            action = np.argmax(q_table[(state[0],state[1])])
            print ("moving ",action)
            next_state, reward, status = env.step(action,state)

        else:
            action = random.randint(0,3)
            next_state, reward, status = env.step(action,state)
        print (itera)
        if (episode==999):
            env2.render(next_state)
        
        cur = (1-alpha) * q_table[state[0]][state[1]][action]
        nex = alpha * (reward + gamma* np.max( q_table[next_state[0]][next_state[1]][:]) )
        
        q_table[state[0]][state[1]][action] = cur + nex
        total_reward += reward
    
        state = next_state
        itera += 1
        print ("Reward -------", reward)
        plt.show()
        
        if state==[4,4]:
            total_reward_history.append(total_reward)
            print (total_reward)
            print ("-"*250)
    
```
