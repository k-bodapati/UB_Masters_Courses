```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torch.tensor as tensor
import numpy as np
import matplotlib.pyplot as plt
from torch.autograd import Variable
import gym
```

```python
env = gym.make("CartPole-v0")
act_space = env.action_space.n
obs_space = env.observation_space.shape[0]
print (act_space, obs_space)
```

```python
class Network(nn.Module):
    
    def __init__(self, input_size, output_size):
        super(Network, self).__init__()
    
        self.layer1 = nn.Linear(input_size, 10, bias=True)
        self.layer2 = nn.Linear(10, output_size, bias=True)
        self.optim = torch.optim.RMSprop(self.parameters(), lr=0.01)
    
    def forward(self, inp):
        x = self.layer1(inp)
        x = F.relu(x)
        x = self.layer2(x)
        x = F.softmax(x)
        return x
```

```python
class Agent():
    def __init__(self):
        self.total_episodes = 4000
        self.gamma = 0.99
        self.max_time = 201

        self.state_mem = list()
        self.action_mem = list()
        self.reward_mem = list()
    
    def update_policy(self, policy, total_steps):
        
        running_add = 0
        for i in reversed(range(total_steps)):
            if self.reward_mem[i] == 0:
                running_add = 0
            else:
                running_add = running_add * self.gamma + self.reward_mem[i]
                self.reward_mem[i] = running_add

        # Normalize reward
        reward_mean = np.mean(self.reward_mem)
        reward_std = np.std(self.reward_mem)
        for i in range(total_steps):
            self.reward_mem[i] = (self.reward_mem[i] - reward_mean) / reward_std

        # Gradient Desent
        policy.optim.zero_grad()

        for i in range(total_steps):
            state = agent.state_mem[i]
            action = Variable(torch.FloatTensor([agent.action_mem[i]]))
            reward = agent.reward_mem[i]

            action_here = torch.distributions.Categorical(policy(state))
            loss = -action_here.log_prob(action) * reward 
            loss.backward()
        policy.optim.step()

        
            

        
```

```python
policy = Network(obs_space, act_space)
agent = Agent()

total_steps = 0
score_list = []

for episode in range(agent.total_episodes):

    state = env.reset()
    

    for step in range(agent.max_time):
        state = torch.from_numpy(state).float()
        total_steps += 1
        action = policy(state)
        action = torch.distributions.Categorical(action)
        action = action.sample()

        next_state, reward, done, info = env.step(action.item())
        
        agent.action_mem.append(action)
        agent.state_mem.append(state)
        agent.reward_mem.append(reward)

        if not done:
            state = next_state
        else:
            reward = 0
            break
    print ("Episode : ", episode, "Steps : ", step)
    score_list.append(step)
    
    if episode > 0 and episode % 5 == 0:
        agent.update_policy(policy, total_steps)
        agent.state_mem.clear()
        agent.action_mem.clear()
        agent.reward_mem.clear()
        total_steps = 0
```

### Advantage Actor Critic

```python
import torch
import torch.nn as nn
import numpy as np
import gym
import matplotlib.pyplot as plt
from torch.autograd import Variable
```

```python
if torch.cuda.is_available():
    device = "cuda:0"
else:
    device = "cpu"
```

```python
class Network(nn.Module):
    def __init__(self, inp_size, out_size):
        super(Network,self).__init__()
        # Actor
        self.a1 = nn.Linear(inp_size, 256)
        self.a2 = nn.Linear(256, out_size)
        # Critic
        self.c1 = nn.Linear(inp_size,256)
        self.c2 = nn.Linear(256, 1)

        self.lr = 3e-4
        self.optim = torch.optim.Adam(self.parameters(), lr=self.lr)
    
    def forward(self, state):
        # state = Variable(torch.from_numpy(state).float().unsqueeze(0))
        action_dist = self.a1(state)
        action_dist = torch.relu(action_dist)
        action_dist = self.a2(action_dist)
        action_dist = nn.functional.softmax(action_dist, dim=0)

        value_est = self.c1(state)
        value_est = torch.relu(value_est)
        value_est = self.c2(value_est)
        
        return (action_dist, value_est)

        
```

```python
class Agent():
    def __init__(self):
        self.gamma = 0.99
        self.episodes = 3000
        self.max_steps = 200
    
    def update_policy(self, value_est_List, reward_list, log_probs_list, Q_value):
        value_est_tensor = torch.tensor(value_est_List).float()
        log_probs_tensor = torch.stack(log_probs_list)
        
        Q_value_array = np.zeros(value_est_tensor.shape)
        
        for i in reversed(range(len(rewards_list))):
            Q_value = reward_list[i] + (self.gamma * Q_value)
            Q_value_array[i] = Q_value
        
        Q_value_tensor = torch.tensor(Q_value_array)
        advantage_tensor = Q_value_tensor - value_est_tensor
        # print (Q_value_tensor)
        actor_loss = (-log_probs_tensor * advantage_tensor).mean()
        critic_loss = 0.5 * advantage_tensor.pow(2).mean()
        total_loss = actor_loss + critic_loss

        net.zero_grad()
        total_loss.backward()
        net.optim.step()

        
```

```python
env = gym.make("CartPole-v0")
act_space = env.action_space.n
obs_space = env.observation_space.shape[0]
agent = Agent()
net = Network(obs_space, act_space)
```

```python
score_list = list()
for episode in range(0,agent.episodes):
    state = env.reset()
    state = torch.tensor(state).float()

    value_est_list = list()
    rewards_list = list()
    log_prob_list = list()
    done = False

    for step in range(0,agent.max_steps):
        action_dist, value_est = net(state)
        value_est_list.append(value_est)

        action = np.random.choice(act_space, p=np.array(action_dist.data))
        log_prob = torch.log(action_dist)[action]
        log_prob_list.append(log_prob)

        next_state, reward, done, info = env.step(action)
        next_state = torch.tensor(next_state).float()
        rewards_list.append(reward)

        if done:
            print ("Episode ", episode, "Total Reward ", step)
            _ , Qval = net.forward(next_state)
            score_list.append(step)
            break
        else:
            state = next_state
        
    
    agent.update_policy(value_est_list, rewards_list,log_prob_list, Qval)

        
        
```
